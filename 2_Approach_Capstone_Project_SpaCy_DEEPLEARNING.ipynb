# SpaCy Named Entities - Fake News

> Indented block



"A named entity is a “real-world object” that’s assigned a name – for example, a person, a country, a product or a book title. spaCy can recognize various types of named entities in a document, by asking the model for a prediction. Because models are statistical and strongly depend on the examples they were trained on, this doesn’t always work perfectly and might need some tuning later, depending on your use case."




1 Fake

```python
text = "New research from Harvard University shows that the chemical in chilli peppers that causes the hot sensation in your mouth reduces the replication rate of coronaviruses. The researchers are currently investigating whether adding more spicy foods to your diet coul dhelp combat COVID-19."

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

# Named entities are available as the ents property of a Doc
displacy.render(doc, style="ent", jupyter=True)
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("New research from Harvard University shows that the chemical in chilli peppers that causes the hot sensation in your mouth reduces the replication rate of coronaviruses. The researchers are currently investigating whether adding more spicy foods to your diet coul dhelp combat COVID-19.")

for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)

displacy.render(doc, jupyter=True,  options={'distance': 70})     
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("New research from Harvard University shows that the chemical in chilli peppers that causes the hot sensation in your mouth reduces the replication rate of coronaviruses. The researchers are currently investigating whether adding more spicy foods to your diet coul dhelp combat COVID-19.")

# document level
ents = [(e.text, e.start_char, e.end_char, e.label_) for e in doc.ents]
print(ents)

# token level
ent_san = [doc[0].text, doc[0].ent_iob_, doc[0].ent_type_]
ent_francisco = [doc[1].text, doc[1].ent_iob_, doc[1].ent_type_]
print(ent_san)  # ['San', 'B', 'GPE']
print(ent_francisco)  # ['Francisco', 'I', 'GPE']

displacy.render(doc, style="ent", jupyter=True)    
```
